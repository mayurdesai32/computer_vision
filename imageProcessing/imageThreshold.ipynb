```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('../data/rainbow.jpg')
plt.imshow(img)
```

```python
# Adding the 0 flag to read it in black and white

img = cv2.imread('../data/rainbow.jpg',0)
plt.imshow(img,cmap='gray')
```

```python
# different Type of Threshold
# 1 binary
# 2 binary Inverse
# 3 Threshold Truncation
# 4 Threshold to Zero
# 5 Threshold to Zero (Inverse)
# 6 Adaptive Threshold
```

```python
# 1 binary Threshold
ret,thresh1=cv2.threshold(img,127,255,cv2.THRESH_BINARY)
plt.imshow(thresh1,cmap='gray')
```

```python
# 2 binary Inverse Threshold
ret,thresh1=cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
plt.imshow(thresh1,cmap='gray')
```

```python
# 3 Threshold Truncation Threshold
ret,thresh1=cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
plt.imshow(thresh1,cmap='gray')
```

```python
#  4 Threshold to Zero Threshold
ret,thresh1=cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
plt.imshow(thresh1,cmap='gray')
```

```python
# 5 Threshold to Zero (Inverse) Threshold
ret,thresh1=cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)
plt.imshow(thresh1,cmap='gray')
```

```python

# 6 Adaptive Threshold
ret,thresh1=cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
plt.imshow(thresh1,cmap='gray')
```

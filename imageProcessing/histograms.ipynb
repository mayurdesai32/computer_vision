```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dark_horse = cv2.imread('../DATA/horse.jpg') #origin bgr
show_horse = cv2.cvtColor(dark_horse, cv2.COLOR_BGR2RGB) #converted to rgb

rainbow = cv2.imread('../DATA/rainbow.jpg')
show_rainbow = cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)

blue_bricks = cv2.imread('../DATA/bricks.jpg')
show_bricks = cv2.cvtColor(blue_bricks, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(show_horse)


```

```python
# blue_bricks in BGR
# mask parameter for part of image 
# histSize parameter for uppervalue
hist_value=cv2.calcHist([blue_bricks],channels=[0],mask=None,histSize=[256],ranges=[0,256])
```

```python
hist_value.shape
```

```python
plt.plot(hist_value)
```

```python
# dark horse
hist_value = cv2.calcHist([dark_horse], channels=[0],
                          mask=None, histSize=[256], ranges=[0, 256])
plt.plot(hist_value)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
backpackImg=cv2.imread('../data/dog_backpack.jpg')
waterImg=cv2.imread('../data/watermark_no_copy.png')
```

```python
plt.imshow(backpackImg)
```

```python
plt.imshow(waterImg)
```

```python
backpackImg=cv2.resize(backpackImg, dsize=(800, 800))
waterImg = cv2.resize(waterImg, dsize=(800, 800))

backpackImg = cv2.cvtColor(backpackImg, cv2.COLOR_BGR2RGB)
waterImg = cv2.cvtColor(waterImg, cv2.COLOR_BGR2RGB)



```

```python
#this is the formula 1+2+
img = cv2.addWeighted(src1=backpackImg, alpha=0.8, src2=waterImg,beta=0.2,gamma=0)

plt.imshow(img)
```

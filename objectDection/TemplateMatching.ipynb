```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
fullImg= cv2.imread('../data/sammy.jpg')
```

```python

fullImg = cv2.cvtColor(fullImg, cv2.COLOR_BGR2RGB)
plt.imshow(fullImg)
```

```python
face = cv2.imread('../DATA/sammy_face.jpg')
face = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)
plt.imshow(face)
```

```python
height, width,channels = face.shape
```

```python
my_method=eval('cv2.TM_CCOEFF')
res = cv2.matchTemplate(fullImg, face,my_method)
plt.imshow(res)
```

```python
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)

top_left=max_loc
bottom_right = (top_left[0]+width, top_left[1]+height)
cv2.rectangle(fullImg, top_left, bottom_right,color=(255,0,0),thickness=10)
plt.subplot(121)
plt.imshow(res)
plt.title('HEATMAP OF TEMPLATE MATCHING')

plt.subplot(122)
plt.imshow(fullImg)
plt.title('DETECTION OF TEMPLATE')
plt.show()
```

```python
# now all the method
```

```python
# All the 6 methods for comparison in a list
# Note how we are using strings, later on we'll use the eval() function to convert to function
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR','cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

```python
for m in methods:
 # creating copy
 fullImg_copy = fullImg.copy()
 method=eval(m)
 # template matching
 res = cv2.matchTemplate(fullImg_copy, face,method)
 min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
 if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
  top_left=min_loc
 else:
  top_left=max_loc
 bottom_right = (top_left[0]+width, top_left[1]+height)
 cv2.rectangle(fullImg_copy, top_left, bottom_right,color=(255,0,0),thickness=10)
  # plot and show the image
 plt.subplot(121)
 plt.imshow(res)
 plt.title('HEATMAP OF TEMPLATE MATCHING')
 plt.subplot(122)
 plt.imshow(fullImg_copy)
 plt.title('DETECTION OF TEMPLATE')
 plt.suptitle(m)
 plt.show()

 print('\n')
 print('\n')
 print('\n')
```

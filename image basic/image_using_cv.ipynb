```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

```

```python
img=cv2.imread('../data/00-puppy.jpg')
# note we have to verify it as it dont throw error

img
# if img file doesnt exit it say None else 
```

```python
plt.imshow(img)
```

```python
type(img)
```

```python
# we got this image because in opencv it read as BGR NOT RGB
# TO Convert int rgb
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python

img.shape
# it will give tuple containing width,height,chanel
```

```python
# for resizing the image
# i.e height=500 and width=1000
img=cv2.resize(img,(1000,500))
plt.imshow(img)
```

```python
# we can also resize the resize by ratio
w_ratio=0.5
h_ratio=0.5
img = cv2.resize(img, (0, 0), img, w_ratio, h_ratio)
plt.imshow(img)
```

```python
# we can flip the image
```

```python
newImg=cv2.flip(img,0)
plt.imshow(newImg)
```

```python
newImg = cv2.flip(img, 1)
plt.imshow(newImg)
```

```python
newImg = cv2.flip(img, -1)
plt.imshow(newImg)
```

```python
# for saving the image
cv2.imwrite('puppyfig.jpg',img)
```

```python
# display lager image in nokebook
fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(111)
ax.imshow(img)
```

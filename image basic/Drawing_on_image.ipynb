```python
import matplotlib.pyplot as plt
import numpy as np 
import cv2
```

```python
blankImg=np.zeros((512,512,3),dtype='int16')
```

```python
plt.imshow(blankImg)
```

```python

cv2.rectangle(blankImg,pt1=(384,100),pt2=(300,200),color=(255,0,0),thickness=5)
# it act as inplace and image will change
```

```python
plt.imshow(blankImg)
```

```python
# for circle
cv2.circle(blankImg,center=(200,300),radius=30,color=(0,255,0),thickness=5)
plt.imshow(blankImg)
```

```python
# for circle fill all color
cv2.circle(blankImg, center=(100, 200), radius=30,
           color=(255, 0, 0), thickness=-1)
plt.imshow(blankImg)
```

```python
cv2.line(blankImg,pt1=(1,10),pt2=(500,480),color=(0,0,2555),thickness=5)

plt.imshow(blankImg)
```

```python
# test
font=cv2.FONT_HERSHEY_COMPLEX
cv2.putText(blankImg,text='hello world',org=(10,400),fontFace=font,fontScale=5,color=(255,0,0))

plt.imshow(blankImg)
```

```python
# polygons
blank_img1 = np.zeros(shape=(512, 512, 3), dtype=np.int32)


```

```python
vertices=np.array([[100,200],[400,300],[300,500],[200,400]],dtype=np.int32)
```

```python
pts = vertices.reshape((-1,1,2))
```

```python
pts
```

```python
cv2.polylines(blank_img1, [pts], isClosed=True, color=(0, 255, 0), thickness=5)
plt.imshow(blank_img1)
```
